```python
cubes = {"red": 0, "green": 1, "blue": 2}

with open("example_input.txt") as f:
    lines = f.readlines()

id_total = 0
for line in lines:
    if not line:
        continue
    game_id, games = line.split(":", 1)
    game_id = game_id.split()[1]
    games = games.split(";")
    for game in games:
        numbers_cubes = game.split()
        for batched in zip(*[iter(numbers_cubes)] * 3):
            cubes = [Cube(*batched, 3)]
            for cube in cubes:
                if cube.number > cube.max:
                    raise ValueError("Cube number is greater than max")
                id_total += cube.number
        break
```
